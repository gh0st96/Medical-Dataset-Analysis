## Hypothesis Testing

Hypothesis 1

Hypothesis: The mean age of patients who survived (death=0) is the same as the mean age of patients who did not survive (death=1).

Null Hypothesis (H0):average age, survivors=average age, non-survivors

Alternative Hypothesis (H1):average age, survivors≠average age, non-survivors


Test Method: Two-sample t-test

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='death', y='age', data=df, palette='Set1')
plt.title('Boxplot of Age for Survivors and Non-Survivors')
plt.xlabel('Survival Status (death)')
plt.ylabel('Age')
plt.show()
```

```python
from scipy.stats import ttest_ind

survivors_age = df[df['death'] == 0]['age']
non_survivors_age = df[df['death'] == 1]['age']

t_stat, p_value = ttest_ind(survivors_age, non_survivors_age, equal_var=False)

print(f'T-test Statistic: {t_stat}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

As this hypothesis test depicts both statistically and visually, we can assume that the avearage age of the surviors is not equal to average age of survivors, which is people who are younger tend to survive

# Hypothesis 2

Hypothesis: There is a significant difference in the mean white blood cell counts (wblc) between patients with and without diabetes.

Null Hypothesis (H0):wblc, diabetes=0=wblc, diabetes=1

Alternative Hypothesis (H1):wblc, diabetes=0≠wblc, diabetes=1



Test Method: Two-sample t-test

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='diabetes', y='wblc', data=df, palette='Set1')
plt.title('Boxplot of White Blood Cell Count for Patients with and without Diabetes')
plt.xlabel('Diabetes Status')
plt.ylabel('White Blood Cell Count (wblc)')
plt.show()
```

```python
diabetes_0_wblc = df[df['diabetes'] == 0]['wblc']
diabetes_1_wblc = df[df['diabetes'] == 1]['wblc']

t_stat, p_value = ttest_ind(diabetes_0_wblc, diabetes_1_wblc, equal_var=False)

print(f'T-test Statistic: {t_stat}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

As the above visualization and T tet depicts, There is a significant difference in the mean white blood cell counts (wblc) between patients with and without diabetes as we fail to reject the null hypothests
